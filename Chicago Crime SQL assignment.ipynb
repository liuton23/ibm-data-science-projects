### Problem 3: List all case numbers for crimes involving minors?

```python
%sql drop view child_crime
%sql create view child_crime as select case_number from crime where primary_type like '%CHILD%' or description like '%CHILD%'
%sql select * from child_crime
```

### Problem 4: List all kidnapping crimes involving a child?

```python
%sql select * from crime, child_crime where crime.case_number = child_crime.case_number and primary_type = 'KIDNAPPING'
```

### Problem 5: What kind of crimes were recorded at schools?

```python
%sql select distinct primary_type from crime where location_description like '%SCHOOL%'
```

### Problem 6: List the average safety score for all types of schools.

```python
%%sql 
select "Elementary, Middle, or High School" as type, avg(safety_score) as avg_score 
from schools 
group by "Elementary, Middle, or High School"
```

### Problem 7: List 5 community areas with highest % of households below poverty line.

```python
%sql select * from census order by percent_households_below_poverty desc limit 5
```

### Problem 8: Which community area(number) is most crime prone?

```python
%%sql 
select community_area_number, count(*) as prone 
from crime 
group by community_area_number
order by prone desc 
limit 1
```

### Problem 9: Use a sub-query to find the name of the community area with highest hardship index.

```python
%sql select community_area_name from census where hardship_index = (select max(hardship_index) from census)
```

### Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%%sql 
select community_area_name 
from census, (select community_area_number, count(*) as num from crime group by community_area_number) sub
where census.community_area_number = sub.community_area_number
order by num desc
limit 1
```
